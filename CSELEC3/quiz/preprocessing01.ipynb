## Handling missing values in `normalized-losses`

```python
df[df['normalized-losses']=='?']
```

```python
df2 = df.replace({'normalized-losses': '?'},np.nan)
df2['normalized-losses'].astype(np.float64)
```

## Visualizing relationship between `symboling` and `normalized-losses`

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(df2.symboling, df2['normalized-losses'])
plt.xlabel('Symboling')
plt.ylabel('Normalized losses')
```

Replace missing values in `normalized-losses` with mean of its `symboling`

```python
sym_mean_losses = dict(df2.groupby('symboling')['normalized-losses'].mean())
```

```python
sym_mean_losses
```

```python
(lambda x: sym_mean_losses[x])(-1)
```

```python
missing_norm_losses = df2['normalized-losses'].isna()
 
arr = np.fromiter(map(lambda x: sym_mean_losses[x], df2[missing_norm_losses]['symboling'].values), dtype='float64')
df2[missing_norm_losses]['normalized-losses'] = arr
```
