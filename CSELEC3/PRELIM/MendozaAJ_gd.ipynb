```python
import pandas as pd
import numpy as np
import math
from sklearn import linear_model
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/ajmdz/CSELEC_2-3/master/CSELEC3/data/scores_sample.csv')
df
```

```python
x = df.math
y = df.cs
```

```python
def gradient_descent(x,y):
    m_curr = b_curr = 0
    prev_cost = 0
    iterations = 500000
    learning_rate = 0.000201
    n = len(x)
    
    
    for i in range(iterations):
        
        try: 
            y_predicted = m_curr * x + b_curr
            cost = (1/n) * sum([val**2 for val in (y-y_predicted)])
            md = -(2/n)*sum(x*(y-y_predicted))
            bd = -(2/n)*sum(y-y_predicted)
            m_curr = m_curr - learning_rate * md        
            b_curr = b_curr - learning_rate * bd
            
        except OverflowError:
            print(f"Overflowed on i={i}")
            return
        
        if math.isclose(prev_cost, cost, rel_tol=1e-20):
            break
        prev_cost = cost
        # print(f"prev_cost = {prev_cost}")
#         print(f'optimal cost ={cost}')
        print(f"cost: {cost}, m: {m_curr}, b: {b_curr}, iteration: {i}")
gradient_descent(x, y)
```
