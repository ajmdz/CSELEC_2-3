Read Excel

```python
df_excel = pd.read_excel("sample_tick_price.xlsx", "Sheet1")
df_excel
```

```python
def convert_peop_cell(cell):
    if cell=="n.a.":
        return 'Michael Casabuena'
    return cell

def convert_price_cell(cell):
    if cell=="n.a.":
        return 55
    return cell

df_excel_conv = pd.read_excel("sample_tick_price.xlsx", "Sheet1", converters= {
    'people': convert_peop_cell,
    'price': convert_price_cell
})

df_excel_conv
```
