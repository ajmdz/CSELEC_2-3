```python
import pandas as pd

df = pd.read_excel('sample_tick_price.xlsx')
df
```

### Skip rows

```python
df = pd.read_csv('sample_tick_price.csv', skiprows=1)
df
```

### No header, add header names

```python
df = pd.read_csv('sample_tick_price.csv', header=None, names=['ticker','eps','revenue','people'])
df
```

### First n rows

```python
df = pd.read_csv('sample_tick_price.csv', nrows=2)
df
```

### Replace values with na

```python
df = pd.read_csv('sample_tick_price.csv', na_values=['n.a.', 'not available'])
df
```

```python
df = pd.read_csv('sample_tick_price.csv', na_values= {
    'eps': ['not available'],
    'revenue': [-1],
    'price': ['n.a.'],
    'people': ['n.a.'],
})
df
```

### Save to file

```python
df.to_csv('new_ticker_data.csv', index=False) # no index
```

```python
df.to_csv('new_ticker_data_no_header.csv', index=False, header=False) # no index & header
```
