Checking for NA values in each column

```python
df.isna().sum()
```

Replacing 0 values in Landsize and BuildingArea with NaN

```python
new_df = df.replace({
    'Landsize': 0,
    'BuildingArea': 0,
}, np.nan)
new_df.dropna(subset=['Landsize', 'BuildingArea'], inplace=True)
new_df.shape
```

Houses in each suburb after dropping NaN/0 values in Landsize and BuildingArea

```python
new_df.groupby('Suburb').size()
```

Median price, land size, and building area in each suburb

```python
new_df.groupby('Suburb')[['Price','Landsize', 'BuildingArea']].median()
```
