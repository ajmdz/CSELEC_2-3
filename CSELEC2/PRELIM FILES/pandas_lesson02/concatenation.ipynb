```python
import pandas as pd

ph_weather = pd.DataFrame({
    "city": ["cebu", "manila", "qc"],
    "temperature": [29, 35, 31],
    "humidity": [80, 60, 78]
})

ph_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["ny", "la", "chicago"],
    "temperature": [21, 35, 14],
    "humidity": [62, 63, 75]
})
us_weather
```

```python
df = pd.concat([ph_weather, us_weather])
df
```

```python
df = pd.concat([ph_weather, us_weather], ignore_index=True)
df
```

```python
df = pd.concat([ph_weather, us_weather], keys=["ph", "us"])
df
```

```python
df.loc["ph"]
```

```python
temperature_df = pd.DataFrame({
    "city": ["cebu", "manila", "qc"],
    "temperature": [29, 35, 31],
}, index=[0,1,2] )

temperature_df
```

```python
windspeed_df = pd.DataFrame({
    "city": ["cebu", "manila"],
    "windspeed": [7, 15],
}, index=[1,0] )

windspeed_df
```

```python
df = pd.concat([temperature_df, windspeed_df], axis=1)
df
```

```python
series_con = pd.Series(["Dry", "Humid", "Rain"], name="event")
series_con
```

```python
df = pd.concat([temperature_df,series_con], axis=1)
df
```
