# Quiz 1 - Data Preprocessing

```python
import pandas as pd
import numpy as np
df = pd.read_excel('ds-quiz_01.xlsx')
df.shape
```

Replace all `?` values with `NaN`

```python
df = df.replace('?', np.nan)
df.isnull().sum()
```

Fill missing values in `num-of-doors` with column mode

```python
doors_mode = df['num-of-doors'].mode()[0]
df = df.fillna({'num-of-doors':doors_mode})
df['num-of-doors'].isnull().sum()
```

Check for duplicate rows (if any) before dropping. There are no duplicates

```python
df[df.duplicated(keep='first') == True]
```

Convert columns to their appropriate type

```python
df.columns[df.isnull().any()].tolist()
```

Convert all the incomplete columns above to float since they are all numerical

```python
inc_numeric_cols = df.columns[df.isnull().any()].tolist()

for col in inc_numeric_cols:
    df[col] = df[col].astype(np.float64)
```

Fill `NA` values with their column mean

```python
for col in inc_numeric_cols:
    col_mean = df[col].mean()
    df[col] = df[col].fillna(col_mean)

df.isnull().sum()
```
